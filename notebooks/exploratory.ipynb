# Exploratory Walkthrough
This notebook demonstrates basic counts, MAF filtering, and mock ClinVar annotations.
> Data are synthetic; this is a demo without PII.

````python
from scripts.utils import open_maybe_gzip, parse_info
vcf_path = 'data/example.vcf.gz'
n = 0
maf_vals = []
with open_maybe_gzip(vcf_path, 'rt') as f:
    for line in f:
        if line.startswith('#'): continue
        n += 1
        info = line.rstrip().split('\t')[7]
        d = parse_info(info)
        if 'MAF' in d:
            maf_vals.append(float(d['MAF']))
n, maf_vals[:5]
````

````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'maf': maf_vals})
df['maf'].plot(kind='hist', bins=10, title='MAF distribution (synthetic)')
plt.xlabel('MAF'); plt.ylabel('Count'); plt.show()
````
